```python
import pandas as pd 

df = pd.read_csv("customer_shopping_behavior.csv")
 
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include = "all")
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
 df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new columns for age_group 

labels = ['young adult','adult','middle adult','senior']
df['age_group'] = pd.qcut(df['age'] , q=4,labels = labels )
```

```python
df[["age","age_group"]].head()
```

```python
# purchased frequency days 
frequency_mapping = {
    'fortnightly' :14,
    'Weekly' : 7,
    'Monthly':30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' :365,
    'Every 3 Months':90
}

df[' purchased_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.columns
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used',axis = 1)
```
